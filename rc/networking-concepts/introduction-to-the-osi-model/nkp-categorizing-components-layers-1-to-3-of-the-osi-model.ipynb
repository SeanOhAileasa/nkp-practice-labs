The lab topology is composed of two subnets separated by an Ubuntu device called ``PLABRTR01`` functioning as a router. 

The first subnet contains three computers. The first, called ``PLABDC01`` is a Windows 2019 Server with an IP address of ``192.168.0.2/24`` and functions as the Domain Controller. - PLABSA01 (Member Server - Windows Server 2019)
- PLABWIN10 (Workstation - Windows 10)

- PLABWEB01 (Web Server - Windows Server 2019)

###### Examine Components Belonging to Physical Layer of the OSI Model
***

Will examine components of the physical layer of the OSI model. Examples of these components include hardware IDs and connection signal detected on a network adapter.

###### Step 1

Connect to the ``PLABWIN10`` device.

Right-click the Network icon at the bottom right and select ``Open Network & Internet Settings``.

###### Step 2

Settings - Status window is displayed.

Select the Ethernet option on the navigation pane at the left.

###### Step 3

On the Settings - Ethernet window displayed, select the Network and Sharing Center link.

###### Step 4

On the Network and Sharing Center window, select the Change adapter settings link at the left.

###### Step 5

Right-click the listed Ethernet connection and select Properties.

###### Step 6

On the Ethernet Properties dialog box, click Configure… under the Microsoft Hyper-V Network Adapter.

###### Step 7

On the Ethernet pop-up click Yes.

###### Step 8

The Microsoft Hyper-V Network Adapter Properties dialog box appears. This dialog box displays properties of the physical Network Interface Card (NIC) on the device.

Note: PLABWIN10 device is a virtual machine (VM) and does not actually have physical hardware associated with it. That is why the adapter is called a Hyper-V Network Adapter. However, it is still considered to be a part of the physical layer of the OSI model.

Access the Details tab.

###### Step 9

From the Property dropbox, select Hardware Ids.

###### Step 10

Notice two values appear in the Value box.

These are hardware IDs for the network adapter being used.

###### Step 11

From the Property dropbox, select Is connected.

###### Step 12

The Value returned is <true>. This indicates that the NIC is physically connected to a switch or other network device and that a carrier signal in the form of a voltage is detected on the network.

###### Examine Components of the Data Link Layer of the OSI Model
***

Will examine components of the Data Link layer of the OSI model. The Data Link layer is composed of two sublayers: the Logical Link Control (LLC) lower sublayer and the Media Access Control (MAC) upper sublayer. Components/protocols relevant to the Data Link layer include driver details, MAC filtering, and ARP tables.

###### Step 1

Connected to the PLABWIN10 device.

On the Microsoft Hyper-V Network Adapter Properties window displayed.

Access the Driver tab.

###### Step 2

On the Driver tab, click the Driver Details button.

###### Step 3

The Driver File Details window appears. Click on the relevant driver file to view information such as file version and provider of the driver.

According to Microsoft, the LLC sublayer is implemented in the driver of the interface. The LLC takes care of the error-free transfer of data frames from one node to the other.

Close all the open windows to access the desktop.

###### Step 4

Click in the Type here to search text box and type cmd.

Press Enter.

###### Step 5

The Command Prompt window appears.

Type the following command and press Enter.

```
ipconfig /all
```

###### Step 6

Scroll up and review the information found under the Ethernet adapter Ethernet: heading. Locate the Physical Address heading and note the value shown.

This is the MAC address of the specific network interface card (NIC).

Note: Each NIC has a unique MAC address.

Note: The format in which a MAC address is displayed depends on the vendor. The official representation should be in hexadecimal where every two digits are separated by a colon. 

For example, the MAC address shown above would be written as ``00:15:5D:60:64:A5``. Different vendors choose to display MAC addresses differently. As seen below, Microsoft uses hyphens instead of semicolons while Cisco uses periods to separate the address into three sections of four hexadecimal digits as ``0015.5D60.64A5``.

Determining the MAC address of a device can enable implementing MAC address filtering. This is done by recording MAC address of the device in question and adding it to a whitelist or a blacklist to either allow or deny access, as required.

###### Step 7

There are several protocols that function at the Data Link layer. One of these is the Address Resolution Protocol (ARP). This protocol determines the MAC address that corresponds to the appropriate IP address.

To view the ARP table of the PLABWIN10 device, enter the following command in the Command Prompt and press Enter.

```
arp -a
```

###### Step 8

Review the ARP entries under the Interface: ``192.168.0.5`` heading. Notice the address ``192.168.0.2``. This is the PLABDC01 device. From this table, you can see the MAC address of the NIC card for this IP address.

May not include the ``192.168.0.2`` entry. If this is the case, attempt to ``ping`` this IP address and issue the ARP command ``arp -a`` once again. The ping should populate the appropriate ARP entry in the table.

###### Examine Components of the Network Layer of the OSI Model
***

Will examine components of the Network layer of the OSI model. These components include the IP address, subnet mask, and default gateway for a device. Configuration details of the IPv4 and IPv6 protocols are a part of the Network layer.

###### Step 1

On the PLABWIN10 device, ensure that the Command Prompt window is displayed.

Type the following command and press Enter:

```
ipconfig /all
```

Scroll up to the Ethernet adapter Ethernet: section and review the information found there.

Notice the following information about the IPv4 protocol:

IPv4 address: ``192.168.0.5``

Subnet Mask: ``255.255.255.0``

Default Gateway: ``192.168.0.250``

Note the following:

- The IPv4 address is a private IP address and can only be used on internal networks (not on the Internet).

- The subnet mask indicates that the range of IP addresses in the same subnet are from ``192.168.0.0`` to ``192.168.0.255``.

- Any destination outside that subnet will require the use of a default gateway. 

- The default gateway is configured as ``192.168.0.250``.

###### Step 2

Notice the following information about the IPv6 protocol:

Link-local IPv6 address: ``fe80::4c97:1de4:1f5c:25e1%5``

This address begins with ``fe80``, so it is only a link-local address - it is self-assigned and not assigned by a DHCP or Stateless Address Auto-configuration (SLAAC) device.

Note: The last two digits on the IPv6 address may be different than shown above. The ``%5`` at the end is a notation used by Microsoft and is called a scope ID.

###### Step 3

Notice the configuration parameters under the Ethernet adapter Ethernet 2 heading.

Now, will change the IPv4 settings on the Ethernet 2 adapter.

Right-click the Network icon on the taskbar and select Open Network & Internet Settings.

###### Step 4

Settings - Status window is displayed.

Select the Ethernet option on the navigation pane at the left.

###### Step 5

On the Settings - Ethernet window displayed, select the Network and Sharing Center link.

###### Step 6

On the Network and Sharing Center window displayed, click the Ethernet 2 link.

###### Step 7

On the Ethernet 2 Status dialog box, click Properties.

###### Step 8

Ethernet 2 Properties dialog box is displayed.

View the listing of items used by the connection.

Notice that both IPv4 and IPv6 are listed.

Select the Internet Protocol Version 4 (TCP/IPv4) item and click Properties.

###### Step 9

The interface is configured to obtain an IPv4 address automatically. However, there is no DHCP server on the network.

Select the Use the following IP address radio button and configure the interface with the following parameters:

IP address: ``147.52.3.17``

Subnet mask: ``255.255.255.128``

Default Gateway: ``147.52.3.1``

Note the following:

- This IP address is what’s known as a public or routable IP address. It can be used on the Internet.

- The subnet mask is different than its classful address. This means that it is using a Variable Length Subnet Mask (VLSM)

Once the configuration is complete, click OK.

###### Step 10

You are presented with a warning concerning multiple default gateways.

Click Yes to continue.

Click OK to close the Ethernet 2 Properties dialog box,

Click Close to close the open dialog box and navigate back to the Network and Sharing Center.

###### Step 11

Restore the Command Prompt window from the taskbar once again.

###### Step 12

Issue the following command and press Enter:

```
ipconfig /all
```

Review the configuration of the interfaces and verify the changes you made are applied.
