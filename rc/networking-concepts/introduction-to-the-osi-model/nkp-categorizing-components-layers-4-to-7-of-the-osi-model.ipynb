###### Examine Components of the Transport Layer of the OSI Model
***

Protocols operating at the Transport layer include netstat and UDP. In this task, you will examine components of the Transport layer (Layer 4) of the OSI model.

###### Step 1

Ensure that all the devices mentioned in the introduction to the exercise are powered on.

Ensure that the Command Prompt window is displayed on the PLABWIN10 device.

Type the following command to clear the screen:

```
cls
```

Next, type the following command and press Enter:

```
netstat
```

###### Step 2

After several seconds, the output appears. This command lists the TCP sessions currently running on the device.
Your output may show very few connections or connections made only to a server called Proxy. This is normal.

After reviewing the listed sessions, open Internet Explorer from the Taskbar.

###### Step 3

Tools and resources intranet page is displayed.

To create a new TCP session, type the following into the address bar and press Enter.

```
https://www.comptia.org
```

###### Step 4

The CompTIA website appears.

###### Step 5

Return to the Command Prompt window and reissue the following command and press Enter.

```
netstat
```

In the result displayed, notice the additional sessions listed. These sessions are all established by the ``192.168.0.5`` IP that represents the PLABWIN10 device. 

Further, all the connections are with the ``Proxy:8080`` device. This is because the devices in the PracticeLabs lab environment and the Internet communicate via the proxy server. Proxy is the domain name of the device, and ``8080`` is the port on which communication takes place.

Notice also that each session has a specific state of ``Established``, ``Time_wait``, and ``Close_wait``. These are different states of a TCP session.

###### Step 6

After several attempts, you may see some entries with UDP in the protocol column. These are not sessions, but UDP datagrams that have been sent or received on the interface. Services that use UDP include DHCP, DNS, and SNMP. You can attempt to repeatedly run the following command which filters out all UDP entries. Press Enter each time to issue the command.

```
netstat -p udp
```

Note that it is unlikely that you will find a UDP netstat entry as very few services using UDP are currently running on the lab topology.

###### Understand Session and Presentation Layers of the OSI Model
***

The Session and Presentation layers of the OSI model are not represented in the TCP/IP model but are included within the Application layer of that model.

The Session layer provides mechanism to open, close, and manage a communication session between end-user applications. Communication sessions consist of requests and responses between applications.

The presentation layer is responsible for formatting and delivering information to the application layer for further processing or display. It relieves the application layer of concern regarding syntactical differences in data representation within the end-user systems.

Examples of these layers functioning within the Windows environment include the use of NetBIOS names for the Session layer and character code translation such as from ASCII to EDCDIC and back.

For the most part these functionalities cannot be directly viewed within a Windows environment as most Windows network components are structured around the TCP/IP model and as such, these layers are included within the Application layer.

###### Examine Components of the Application Layer of the OSI Model
***

###### Step 1

On the PLABWIN10 device, restore the Settings - Ethernet window.

Click the Windows Firewall link.

###### Step 2

Windows Security - Firewall & network protection window is displayed.

Click the Advanced settings link.

Click Yes to the User Account Control pop-up.

###### Step 3

Windows Defender Firewall with Advanced Security icon appears on the taskbar.

Select this icon to display the window.

###### Step 4

Windows Defender Firewall with Advanced Security window is displayed.

Click on the Inbound Rules node in the left pane.

View the middle pane that lists the Inbound Rules.

Look for the following application layer protocols:

- Core Networking - Dynamic Host Configuration Protocol (DHCP-In)

- Remote Desktop - User Mode

- SNMP

- Cast to Device streaming server (HTTP-Streaming-In)

These are network-aware applications that service the user applications utilized by the person operating the device.

###### Step 5

To view details of these processes, access the Task Manager.

Right-click the taskbar and select Task Manager.

###### Step 6

The Task Manager window appears.

Click the More details arrowhead at the bottom.

###### Step 7

On the Processes pane displayed, scroll down to access the Windows Processes section.

On this section, find the Service Host process. Notice that this process is running for the DNS Client, Network Location Awareness, as well as the Hyper-V Remote Desktop Virtualization Service. All these are network applications and services that require this process to run.

This process runs just below the network-aware applications such as web browsers, email clients, Remote Desktop clients, and other such software packages. This is the Application layer of the OSI model in action.
