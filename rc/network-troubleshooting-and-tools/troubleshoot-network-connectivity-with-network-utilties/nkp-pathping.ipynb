###### Lab Topology
***

The lab topology is composed of two subnets separated by an Ubuntu device called ``PLABRTR01`` functioning as a router. 

The first subnet contains three computers. The first, called ``PLABDC01`` is a Windows 2019 Server with an IP address of ``192.168.0.2/24`` and functions as the Domain Controller. The second, called ``PLABDM01`` is also a Windows 2019 Server with an IP address of ``192.168.0.3/24``. The third, called ``PLABSA01`` is a Windows 2019 Server with an IP address of ``192.168.0.4/24``. The forth ``PLABWIN10`` is a Windows 10 device with an IP address of ``192.168.0.5/24``. 

The ``PLABRTR01`` device has an IP address of ``192.168.0.1/24`` on the interface that connects to this subnet. The other interface connects to the second subnet with an IP address of ``192.168.1.254/24``.

This second subnet contains only one other device, the ``PLABWEB01`` device which is also running Windows 2019 Server and has an IP address of ``192.168.1.1/24``.

- PLABDC01 (Domain Controller - Windows Server 2019)
- PLABDM01 (Domain Member - Windows Server 2019)
- PLABSA01 (Member Server - Windows Server 2019)
- PLABWIN10 (Workstation - Windows 10)
- PLABRTR01 (Router - Linux Ubuntu 19.04)
- PLABWEB01 (Web Server - Windows Server 2019)

### Using pathping
***

``pathping`` is a ``Windows-based`` ``TCP/IP`` diagnostic tool that provides information about the network path, network delays and dropped packets from the source to the destination network. 

This tool combines the functionality of ``tracert`` and ``ping`` commands to verify connectivity between devices.

***
###### Test Host Reachability Using Pathping
***

Connect to ``PLABWIN10``.

```
cmd
```

```
pathping 192.168.0.2
```

This command traces the route and computes the network statistics from the source to destination network.

After a few seconds, will get the information about network route taken and statistics on the network latency and packet loss.

Sending a ``pathping`` using a target host’s ``Fully Qualified Domain Name`` - ``FQDN``:

```
pathping plabdc01.practicelabs.com
```

Get information about the network path taken and statistics of the network latency and packet loss.

Test the route to a remote web site like google’s DNS server at ``8.8.8.8``:

```
pathping 8.8.8.8
```

The Practice Lab devices are running behind a firewall and will be unable to trace such routes.

The ``PLABWIN10`` attempts to find the route starting from its IP address ``192.168.0.5``, then on to ``192.168.0.250`` which is the default gateway being used by the ``PLABWIN10`` device.

Initially, ``pathping`` will report that destination host is unreachable because of network security policies on the corporate network and the target network attempt to trace the route from the source to destination network will fail.

On the Computing statistics section, notice the network information collected - the packet went back to the sending node and was unable to reach its destination.
