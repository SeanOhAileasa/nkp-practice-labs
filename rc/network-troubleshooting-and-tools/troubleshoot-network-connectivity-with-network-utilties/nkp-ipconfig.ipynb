###### Lab Topology
***

The lab topology is composed of two subnets separated by an Ubuntu device called ``PLABRTR01`` functioning as a router. 

The first subnet contains three computers. The first, called ``PLABDC01`` is a Windows 2019 Server with an IP address of ``192.168.0.2/24`` and functions as the Domain Controller. The second, called ``PLABDM01`` is also a Windows 2019 Server with an IP address of ``192.168.0.3/24``. The third, called ``PLABSA01`` is a Windows 2019 Server with an IP address of ``192.168.0.4/24``. The forth ``PLABWIN10`` is a Windows 10 device with an IP address of ``192.168.0.5/24``. 

The ``PLABRTR01`` device has an IP address of ``192.168.0.1/24`` on the interface that connects to this subnet. The other interface connects to the second subnet with an IP address of ``192.168.1.254/24``.

This second subnet contains only one other device, the ``PLABWEB01`` device which is also running Windows 2019 Server and has an IP address of ``192.168.1.1/24``.

- PLABDC01 (Domain Controller - Windows Server 2019)
- PLABDM01 (Domain Member - Windows Server 2019)
- PLABSA01 (Member Server - Windows Server 2019)
- PLABWIN10 (Workstation - Windows 10)
- PLABRTR01 (Router - Linux Ubuntu 19.04)
- PLABWEB01 (Web Server - Windows Server 2019)

### Using ipconfig
***

``ipconfig`` - ``Internet Protocol Configuration``, is a tool to display all existing IP network configuration values. 

The tool is also used to change settings of:

- ``Dynamic Host Configuration Protocol`` (``DHCP``) and;


- ``Domain Name System`` (``DNS``)

The ``ipconfig`` command with the switch ``/all`` provides more detailed information than ``ipconfig``.

The parameter ``/release`` releases the IP address for the specified adapter from the ``DHCP`` server. 

The parameter ``/renew`` renews the IP address for the specified adapter from the ``DHCP`` server.

<a id="usetheipconfigTool"></a>

###### Use the Tool
***

Connect to the ``PLABWIN10`` device.

``cmd``

Get the IP information about this computer and save it as a text file:

```
ipconfig /all > ip.txt
```

Read the contents of ``ip.txt`` in Notepad:

```
notepad ip.txt
```

The Notepad application window opens, indicating the ip settings of ``PLABWIN10``.

###### Modify Network Settings 
***

Can disable or uninstall some network settings like ``IPv6`` which will alter the output of the ``ipconfig`` command.

On ``PLABWIN10`` device, click in the ``Type here to search`` text box:

```
control panel
```

Select the ``View network status and tasks`` link under the ``Network and Internet`` heading.

On the ``Network and Sharing Center`` window, under the ``View your active networks`` section, click the ``Ethernet`` link.

On the ``Ethernet Status`` dialog box, click ``Properties``.

On the ``Ethernet Properties`` dialog box, clear ``Internet Protocol Version 6 (TCP/IPv6)`` check box.

```
cmd
```

Get the updated IP settings:

```
ipconfig
```

Notice that the ``Ethernet adapter Ethernet`` section no longer has a ``Link-local IPv6 Address`` as it did in the ``ip.txt`` file.

Cleared the TCP/IPv6 check box for the Ethernet network connection and so all IPv6 parameters have been removed.
